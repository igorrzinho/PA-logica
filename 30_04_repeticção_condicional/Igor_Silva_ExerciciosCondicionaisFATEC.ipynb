## **Exemplo 01**
A empresa *XSoftwares* concedeu um bônus de 20% do valor do salário a todos os
funcionários com tempo de trabalho na empresa igual ou superior a cinco anos, e
10% aos demais. Calcular e exibir o valor do bônus.
Para resolver o problema é necessário conhecer o valor do salário e o tempo
de serviço do funcionário.

### **Algoritmo em Visualg**
image.png

```python
# ENTRADA DE DADOS E ATRIBUIÇÃO PARA AS RESPECTIVAS VARIÁVEIS
salario = float(input("Digite o salário do funcionário:"))
tempo = int(input("Digite a quantidade de horas trabalhadas:"))

#PROCESSAMENTO
if(tempo>=5):
  bonus = float(salario*0.20)
else:
  bonus = float(salario*0.10)

#SAIDA
print("O valor do bonus é de R$",round(bonus,2))
```

## **Exemplo 02**
Sabemos que um aluno é aprovado caso apresente média maior ou igual a 7.0 e
frequência maior ou igual a 75%. Na verdade, em uma situação real, se o aluno
obtiver a frequência mínima exigida e uma média entre 3 e 7, ainda teria direito a uma última avaliação de recuperação.

### **Algoritmo em Visualg**
image.png

```python
# SOLUÇÃO 01
media=float(input("Digite a média do aluno:"))
frequencia=float(input("Digite a frequencia do aluno:"))

if(frequencia>=75):
  if(media>=7):
    print("APROVADO")
  else:
    if(media>=3):
      print("RECUPERAÇÃO")
    else:
      print("REPROVADO POR MÉDIA")
else:
  print("REPROVADO POR FALTAS")
```

```python
# SOLUÇÃO 02
media=float(input("Digite a média do aluno:"))
frequencia=float(input("Digite a frequencia do aluno:"))

if(frequencia>=75 and media>=7):
  print("APROVADO")
elif(frequencia>=75 and media<3):
  print("REPROVADO POR MÉDIA")
elif(frequencia>=75 and media>=3 and media<7):
  print("RECUPERAÇÃO")
else:
  print("REPROVADO POR FALTAS")
```

3. A nota final de um estudante é calculada a partir de três notas atribuídas
respectivamente a um trabalho de laboratório, a uma avaliação semestral e a um
exame final. A média das três notas mencionadas anteriormente obedece aos
pesos a seguir:
 -      |--------------------------------|
 -     | nota                    | peso |
 -     | trabalho de laboratorio | 2    |
 -     | avaliação semestral     | 3    |
 -     | exame final             | 5    |
 -     |--------------------------------|

- Faça um programa que receba as três notas, calcule e mostre a média ponderada e
o conceito que segue a tabela abaixo:

  -      |----------------------------|
  -      | media ponderada | conceito |
  -      | 8 a 10          | a        |
  -      | 7 a 8           | b        |
  -      | 6 a 7           | c        |
  -      | 5 a 6           | d        |
  -      | 0 a 5           | e        |
  -      |----------------------------|


```python
n1 = 2*float(input("Digite a nota tirada no trabalho de laboratorio: "))
n2 = 3*float(input("Digite a nota tirada no avaliação semestral: "))
n3 = 5*float(input("Digite a nota tirada no exame final: "))
media = (n1 + n2 + n3) /10

if (media <= 5):
  nota = "E"
elif (media > 5 and media <= 6):
  nota = "D"
elif(media > 6 and media <= 7):
  nota = "C"
elif(media > 7 and media <= 8):
  nota = "B"
else:
  nota = "A"

print("Sua nota é " , nota)
```

4. Faça um programa que recebe três notas de um aluno, calcule e mostre a média
 aritmética e a mensagem que segue a tabela abaixo. Para alunos de exame,
 calcule e mostre a nota que deverá ser tirada no exame para aprovação,
 considerando que a média no exame é 6,0.
  -      |------------------------------|
  -      | media aritmerica | Mensagem  |
  -      | 0,0 a 3,0        | reprovado |
  -      | 3,0 a 7,0        | Exame     |
  -      | 7,0 a 10,0       | Aprovado  |
  -      |------------------------------|

```python
n1 = float(input("Primeira nota: "))
n2 = float(input("Segunda nota: "))
n3 = float(input("terceira nota: "))
media = (n1 + n2 + n3)/3
if (media < 3):
  sit = "Reprovado"
elif (media >= 3 and media <7):
  print("Voce esta de exame")
  n4 = float(input("Qual nota voce tirou no exame?"))
  media = (media + n4) /2
  if (media > 7):
    sit = "Aprovado"
  else:
    aux = 7 - media
    print("Nao aprovado por",aux,"pontos")
    sit = "Reprovado"
else:
  sit = "Aprovado"

print(sit)
```

5. Uma livraria está fazendo uma promoção para pagamento à vista em que o
  comprador pode escolher entre dois critérios de desconto:
  -  Critério A: R$ 0,25 por livro + R$ 7,50 fixo.
  - Critério B: R$ 0,50 por livro + R$ 2,50 fixo.

  Faça um programa em que o usuário digite a quantidade de livros que deseja
  comprar e o programa diga qual é a melhor opção de desconto.



```python
qtd_livros = int(input("Digite a quantidade de livros que deseja comprar: "))
a = (qtd_livros * 0.25) + 7.50
b = (qtd_livros * 0.50) + 2.50

if (a == b):
  print("Ambas as opções saem o mesmo valor")
elif (a < b):
  print("A melhor opção é: A")
else:
  print("A melhor opção é: B")
```

6.  O posto de atendimento médico e hospitalar de uma pequena cidade atende em
média 138 pessoas por dia e vem observando, ao longo do tempo, que os casos
de pessoas com problemas de sobrepeso vêm aumentando a cada ano. Assim, o
posto de atendimento determinou que em todos os atendimentos o médico ou o
enfermeiro deverão calcular o peso ideal de todos os pacientes atendidos.
Elabore um programa que verifique se o paciente está acima de seu peso ideal
de acordo com as condições abaixo:
  - Para homens: (72,7 altura) – 58;
  - Para mulheres: (62,1 altura) – 44,7.

```python
s = input("Digite o sexo masculino(M), femenino(F)")
alt = float(input("Digite Sua altura: "))
peso = float(input("Digite seu peso em kg: "))

if(s == "m" or s == "M"):
  ideal = (72.7 * alt)-58
  if (peso > ideal):
    print("Voce esta acima do peso ideal de:",ideal,"kg")
  else:
    print("Voce não esta acima do peso ideal de:",ideal,"kg")
elif(s == "f" or s == "F"):
  ideal = (62.1 * alt)-44.7
  if (peso > ideal):
    print("Voce esta acima do peso ideal de:",ideal,"kg")
  else:
    print("Voce não esta acima do peso ideal de:",ideal,"kg")
else:
  print("Digite um valor valido para sexo M ou F")
```
