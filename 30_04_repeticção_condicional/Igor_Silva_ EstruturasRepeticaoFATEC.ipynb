# **ESTRUTURAS DE REPETIÇÃO**

Uma estrutura de repetição em Python é um recurso para desenvolver tarefas repetitivas em um loop contínuo. O loop funciona até uma condição ser satisfeita.

Uma estrutura de repetição em Python funciona da mesma maneira que uma estrutura em outras linguagens, como C e Java. A diferença básica é a sintaxe.

# **ESTRUTURA FOR**

# Exemplo 01 - Utilizando Listas

````python
# Criando uma lista (semelhante a um vetor)
lojas = ['Rio de Janeiro','São Paulo','Belo Horizonte','Curitiba']
````

````python
# Acessando uma determinada posição da lista lojas
lojas[1]
````

**Lembre-se: Listas possuem índice de suas posições começando em 0**

````python
# Exemplo de um laço de repetição para contar de 1 a 10
for i in range(1,11):
  print(i)
````

````python
# Motrando todas as lojas da lista
for loja in lojas:
  print("Loja: ",loja)
````

````python
# Ordenando as lojas
for i in range(4):
  print(i,'- Loja:',lojas[i])
````

````python
# Mas e se não soubessémos o tamanho da nossa lista?
tamanho_lista = len(lojas)
tamanho_lista
````

````python
# Ordenando as lojas
for i in range(tamanho_lista):
  print(i,'- Loja:',lojas[i])
````

Mas e se quissémos exibir:

1ª Loja: Rio de Janeiro

2ª Loja: São Paulo

3ª Loja: Belo Horizonte

4ª Loja: Curitiba

````python
# Ordenando as lojas
for i in range(tamanho_lista):
  i+=1
  print(i,'ª - Loja:',lojas[i])
````

*Quando pedimos para incrementar i dentro laço observe que a 1ª loja lista está incorrreta e ainda causamos um erro de indice dentro da lista.*

````python
# Ordenando as lojas
for i in range(tamanho_lista):
  print(i+1,'ª - Loja:',lojas[i])
````

A estrutura de repetição, também pode ser usada para percorrer caracteres dentro de uma string.

````python
for x in "FATEC MAUÁ":
  print(x)
````

# **ESTRUTURA WHILE**

O comando while faz com que um conjunto de instruções seja executado enquanto uma condição é atendida. Quando o resultado dessa condição passa a ser falso, a execução do loop é interrompida

````python
# Exemplo While
numero = 0
while numero >= 0:
  numero = int(input("Digite um número positivo"))
print('Você digitou um número inválido')
````

````python
# Contador de 1 a 10
num = 1
while num <=10:
  print(num)
  num+=1
````

````python
print("deseja ver uma tabuada? ")
res = input("")
while (res == "s" or res == "S" ):
  num = int(input("qual numero? "))

  for i in range(1,10):
    print(i, " x ", num," = ",num * i)
  print("deseja ver uma tabuada? ")
  res = input("")
print("fim")
````

5) Recibo de Mercado
Faça um programa para um caixa de mercado que imprima um recibo de compras. O programa deve receber o nome de N produtos que estão sendo comprados (N simboliza uma quantidade variada). Utilize as estruturas de repetição.

Também deve receber as informações de quantidade e valor unitário de cada um destes produtos.

Exiba o valor total de cada produto comprado e o valor total das compras de todos os produtos.

*Dica: Faça seu código funcionar com um produto. Quando funcionar com um produto, modifique-o para calcular cinco produtos.*

````python
produtos = []
preco_total = 0
print("adicionar um produto? [S-N]")
res = input("")
while (res == "s" or res == "S" ):
  nome = input("qual nome do produto? ")
  preco = float(input("qual preço do produto? R$"))
  num = int(input("qual a quantidade do produto? "))
  preco_total = preco_total + (preco * num)
  prod = [nome, preco, num, (num * preco) ]

  produtos.append(prod)

  print("adicionar um produto? [S-N]")
  res = input("")
print("Recibo de Compras")
for i in produtos:
  print('produto:',i[0],'- R$',i[1], '- qtd:', i[2],' Total: R$', i[3] )
print("|----------|   Valor Final: R$", preco_total,"   |----------|")
print("|----------------------------------------------------------|")
````

6) Verificador de Números Primos
Escreva um código que recebe um número e imprime uma mensagem dizendo se ele é ou não é um número primo.

**Nota**: *Um número primo é um número que só é divisível (divisão exata / sem resto) por 1 e ele mesmo.*

**Nota²:** *Veja um vídeo do [Khan Academy sobre Números Primos](https://pt.khanacademy.org/math/pre-algebra/pre-algebra-factors-multiples/pre-algebra-prime-numbers/v/prime-numbers).*

````python
n1 = int(input("Digite um numero inteiro: "))
primo = "é um numero primo"
for i in range(2,n1):
  res = n1 % i
  if (res == 0):
    primo = "não é um numero primo"
print(n1, primo)
````
